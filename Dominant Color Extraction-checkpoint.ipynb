```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
from sklearn.cluster import KMeans
```

```python
img = cv2.imread("sceneary.jpeg")
img = cv2.resize(img,(306,382))
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
print(img.shape)
```

```python
plt.imshow(img)
```

```python
#reshaping image from RxCx3 to Mx3

X= img.reshape((-1,3))
print(X.shape)
```

```python
k=9
model = KMeans(n_clusters=k)
model.fit(X)
```

```python
centroids = model.cluster_centers_
```

```python
print(centroids)
```

```python
colors = np.array(centroids,dtype='uint8')
```

```python
print(colors)
```

```python
i=1
for color in colors:
    plt.subplot(1,k,i)
    plt.axis("off")
    i = i+1
    mat = np.zeros((100,100,3), dtype='uint8')
    mat[:,:,:] = color
    plt.imshow(mat)
plt.show()    
```
